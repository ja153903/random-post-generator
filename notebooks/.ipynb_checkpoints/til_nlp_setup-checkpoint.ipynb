```python
import pandas as pd
import numpy as np
import markovify
import re
import spacy
```

```python
til_df = pd.read_csv('./../data/til_data.csv')
```

```python
til_df.columns[0]
```

```python
til_df.drop('0', inplace=True, axis=1)
```

```python
til_df.columns = ['TIL Sentence']
```

```python
til_df['TIL Sentence'].values
```

```python
# create an array to join words as a large string
sentences = " ".join([sentence for sentence in til_df['TIL Sentence'].values])
```

```python
sentences
```

```python
# create markov model
til_text_model = markovify.Text(sentences)
```

```python
til_text_model.make_sentence()
```

```python
for i in range(5):
    print(til_text_model.make_sentence())
```
